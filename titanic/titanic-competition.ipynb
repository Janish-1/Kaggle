```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import pickle
import joblib
```

```python
# Importing data and Viewing
train_data = pd.read_csv("train.csv")
test_data = pd.read_csv("test.csv")
gender_submission = pd.read_csv("gender_submission.csv")
```

```python
# Training Data
train_data.head(10)
```

```python
# Test Data
test_data.head(10)
```

```python
# Gender Submission
gender_submission.head(10)
```

```python
# Refining Training Data
columns_to_drop = ["Name","Ticket","Cabin"]
dropped_train_df = train_data
dropped_train_df = dropped_train_df.drop(columns=columns_to_drop)
dropped_train_df.head(5)
```

```python
# Refining Testing Data
columns_to_drop = ["Name","Ticket","Cabin"]
dropped_test_df = test_data
dropped_test_df = dropped_test_df.drop(columns=columns_to_drop)
dropped_test_df.head(5)
```

```python
# Label Encoding - Trained Data'
label_encoder = LabelEncoder()
columns = ["Sex","Embarked"]
x = dropped_train_df
for i in columns:
    name = i
    x[name] = label_encoder.fit_transform(x[name])
```

```python
# Label Encoded - Trained Data
x.head(5)
```

```python
# Label Encoding - Testing Data
label_encoder = LabelEncoder()
columns = ["Sex","Embarked"]
y = dropped_test_df
for i in columns:
    name = i
    y[name] = label_encoder.fit_transform(y[name])
```

```python
# Label Encoding - Testing Data
y.head(5)
```

```python
# One Hot Encoding - Trained Data
columns = ["Sex","Embarked"]
a = dropped_train_df
for i in columns:
    name = i
    a = pd.get_dummies(a, columns=[name], prefix=[name])
```

```python
# One Hot Encoding - Trained Data
a = a.drop(columns=["Embarked_3"])
a.head(5)
```

```python
# One Hot Encoding - Testing Data
columns = ["Sex","Embarked"]
b = dropped_test_df
for i in columns:
    name = i
    b = pd.get_dummies(b, columns=[name], prefix=[name])
```

```python
# One Hot Encoding - Testing Data
b.head(5)
```

```python
print(x.isnull().sum())
print("---------------")
print(y.isnull().sum())
print("---------------")
print(a.isnull().sum())
print("---------------")
print(b.isnull().sum())
```

```python
x.fillna(-1,inplace=True)
y.fillna(-1,inplace=True)
a.fillna(-1,inplace=True)
b.fillna(-1,inplace=True)
```

```python
# Label Encoding - Model
X = x.drop(columns=["Survived"])
Y = x["Survived"]

# Split the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Check shapes
print("X_train shape:", X_train.shape)
print("Y_train shape:", Y_train.shape)

# Initialize and train the logistic regression model
model = LogisticRegression()

try:
    model.fit(X_train, Y_train)
except ValueError as e:
    print("Error:", e)

# Make predictions
label_encoded_predictions = model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(Y_test, label_encoded_predictions)
print("Accuracy: ", accuracy)

with open('model_titanic_label_encoded.pkl', 'wb') as file:
    pickle.dump(model, file)
```

```python
# One Hot Encoding - Model
X = a.drop(columns=["Survived"])
Y = a["Survived"]

# Split the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Check shapes
print("X_train shape:", X_train.shape)
print("Y_train shape:", Y_train.shape)

# Initialize and train the logistic regression model
model = LogisticRegression()

try:
    model.fit(X_train, Y_train)
except ValueError as e:
    print("Error:", e)

# Make predictions
one_hot_encoded_predictions = model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(Y_test, one_hot_encoded_predictions)
print("Accuracy: ", accuracy)

with open('model_titanic_one_hot_encoded.pkl', 'wb') as file:
    pickle.dump(model, file)
```
